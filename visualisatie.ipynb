```python
column = pd.concat([train['name'], train['price']], axis=1)
column
```

```python
train['category_name'] = train['category_name'].fillna('undefined').astype(str)
train['brand_name'] = train['brand_name'].fillna('undefined').astype(str)
train['shipping'] = train['shipping'].astype(str)
train['item_condition_id'] = train['item_condition_id'].astype(str)
train['item_description'] = train['item_description'].fillna('undefined')
```

```python
#type(train['category_name'].iloc[0])
#pandas.Series.str.split(train['category_name'] pat='/', n=-1, expand=True)
column_split = lambda x: pd.Series([i for i in (x.split('/'))])
splitted = train['category_name'].apply(column_split)
column = pd.concat([train, splitted], axis=1)
print(column.head)
```

```python
data = column
main_categories = set(data[0])
print(main_categories)
```
